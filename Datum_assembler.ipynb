Desmos Abstract Turing Machine (DATUM)
An abstract computer in Desmos
DEFINITIONS
= : value & : address
INSTRUCTION|NUMBER|ARGS
NOP 1
SET 2 (=,&)
HALT 3
(arithmetic, takes 3 addresses and operates on them. Later I might add instructions that can take an immediate value as the second operand )
ADD 4 (in &,in &, out &)
SUB 5 (in &,in &, out &)
MULT 6 (in &,in &, out &)
DIV 7 (in &,in &, out &)
MOD 8 (in &[dividend],in &[divisor], out &)
POW 9 (in &,in &[exponent], out &)
LOG 10 (in &,in &[base], out &)
COPY 11 (in &, out &)
JMP 12 (new PC = ,new PC &,=mode)[the maximum of argument 1 and 3 is chosen except  if the second argument is zero, in which case argument 1 is chosen)
[Jump modes 0:Unconditional 1:Zero 2:Negative]
ALOC 13 (=,&) [the maximum is chosen except  if the second argument is zero, in which case argument 1 is chosen]
14-20 are versions of the arithmetic functions that take an immediate value as the second operand. These are represented with the standard opcode followed by I in assembly 
21-24  are versions of the arithmetic functions 7-10 that take an immediate value as the first operand. These are represented with the standard opcode preceded by I in assembly
ABS 25|
SIN 26 |
COS 27|
REAL 28|
IMAG 29|
FACT 30|
(=,&,out &) [the maximum is chosen except  if the second argument is zero, in which case argument 1 is chosen]

FLAGS Zero:lower bit Negative: upper bit at index 2
Program Counter is at index 1, and increments by 4 after an execution cycle if a jump is not triggered

Input and Output will be handled by writing and reading to and from specific addresses inside the machine that are read or written to by the loop into and out of functions external to the machine.
Instructions are spaced every 4 entries, with the first instruction occurring at index 4

What it does: Parses simple DATUM assembly into 4-word instructions and outputs a LaTeX array in the form shown in your notebook example (e.g. [A\to\left[4,0,0,4,...\right]]).

Argument syntax: Use =n for immediate values, &n for addresses, or plain integers. Each instruction becomes [opcode,a,b,c] (padded with zeros).

Run (CLI):

Pipe assembly on stdin:
printf "ADD &20 &21 &22\nNOP\nHALT\n" | python3 scripts/assembler.py
Or from file:
python3 [assembler.py](http://_vscodecontentref_/2) myprog.asm


```python
from scripts import assembler
assembler.largest_index=60  # Ensure minimum output length
code ="""
# Write assembly code here
NOP
SET 10 49
SET 1 50
SUB 49 50 49
ALOC 1 
JMP 32 0 1
JMP 16 0 0
HALT
"""
print(assembler.assemble_to_latex(code))
```
